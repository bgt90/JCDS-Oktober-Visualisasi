# Marker

```python
jakarta_map=folium.Map(location=[-6.2088, 106.8456],zoom_start=10)

#beri marker di posisi monas

folium.Marker([-6.175392, 106.827153],
             popup='<i>Monas<i>', #kalau diklik keluar tulisan ini
             tooltip='Ini Monumen Nasional', # kalau di hover over keluar tulisan ini
             icon= folium.Icon(color='purple', icon='info-sign')
             ).add_to(jakarta_map)

#kota tua
folium.Marker([-6.1376, 106.8171],
              popup='Kota Tua Jakarta',
              tooltip='Mari ke Kota Tua').add_to(jakarta_map)
# blok M
folium.CircleMarker([-6.2446, 106.8006],
                   radius=30,
                   popup='Blok M',
                   tooltip='Ayo ke sini setelah pandemi berakhir',
                   color='yellow').add_to(jakarta_map)

jakarta_map
```

# Choropleth
Seperti heatmap, choropleth adalah peta yang menunjukkan nilai dari suatu fenomena di suatu wilayah menggunakan warna

Data Source:   
* Jakarta Open Data, https://data.jakarta.go.id/dataset/pelaporan-perpindahan-penduduk-keluar-provinsi-dki-jakarta-tahun-2020-berdasarkan-kelurahan/resource/134253ff-2915-44b8-a1b3-2b1869366ed0  
* Jakarta by Kelurahan geo public, https://pstyd.carto.com/tables/jakarta_by_kelurahan_geo/public 

```python
df=pd.read_csv('pendatang_DKI_Maret_2020.csv')
df.head()
```

```python
# karena data geojson kita per kelurahan, mari kita ambil data pendatang per kelurahan
df_kel=df[['kelurahan','jumlah']]
df_kel
```

```python
df_kel['kelurahan'].nunique()
```

```python
import json
jakarta_by_kel=json.load(open('jakarta_by_kelurahan_geo.geojson'))
# json adalah data javascript yg bentuknya mirip dictionary python
# geojson adalah data geografis yang disimpan dalam bentuk json
```

```python
jakarta_by_kel['features'][3]['properties']['name']
```

```python
jakarta_by_kel['features'][0]['properties']['name']
```

```python
nama=[]
for i in range(len(jakarta_by_kel['features'])):
    nama.append(jakarta_by_kel['features'][i]['properties']['name'])
```

```python
jakarta=r'jakarta_by_kelurahan_geo.geojson'
jakarta_location=[-6.2088, 106.8456]


jakarta_map=folium.Map(jakarta_location,zoom_start=12)

folium.Choropleth(geo_data=jakarta, #data geojson
                 data=df_kel, #data untuk warna choropleth
                 columns=['kelurahan','jumlah'], #kolom df
                 key_on='feature.properties.name', #posisi nama wilayah di geojson
                 fill_color='YlOrRd', #warna
                 fill_opacity=0.7, #transparasi
                 line_opacity=0.2, #transparasi garis boundary
                 legend_name='Jumlah Pendatang di Jakarta pada Maret 2020' #legend
                 ).add_to(jakarta_map)

jakarta_map
```

```python
bins=list(df_kel['jumlah'].quantile([0,0.25,0.5,0.75,1])) #untuk range warna di legend, sekarang jadi berdasarkan quantile
jakarta=r'jakarta_by_kelurahan_geo.geojson'
jakarta_location=[-6.2088, 106.8456]


jakarta_map=folium.Map(jakarta_location,zoom_start=12)

folium.Choropleth(geo_data=jakarta, #data geojson
                 data=df_kel, #data untuk warna choropleth
                 columns=['kelurahan','jumlah'], #kolom df
                 key_on='feature.properties.name', #posisi nama wilayah di geojson
                 fill_color='BuPu', #warna
                 fill_opacity=0.7, #transparasi
                 line_opacity=0.2, #transparasi garis boundary
                 legend_name='Jumlah Pendatang di Jakarta pada Maret 2020', #legend
                 bins= bins
                 ).add_to(jakarta_map)

jakarta_map
```
