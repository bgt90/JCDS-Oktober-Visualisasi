# Matplotlib 
Untuk visualisasi  
inspired by matlab  
Basis untuk library visualisasi di seaborn dan pandas

Matplotlib punya 3 layers untuk coding:
* layer backend
* layer artist --> lebih fleksibel
* layer scripting --> untuk plot sederhana, cara coding juga paling simple

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# %matplotlib inline 
# kalau gambar tidak keluar di notebook
```

```python
# buat data dummy
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

## Scripting Layer

```python
#basic plotting
plt.figure(figsize=(12,7)) #mengatur ukuran gambar
plt.plot(x,y,'r') # plotting garis x dibandingkan dengan y, diberi warna merah
plt.xlabel('ini x', size=15) #membuat label axis horizontal
plt.xticks([1,2,3,4,5],['Andi', 'Budi','Caca','Dedi','Elisa']) 
#mengatur value axis x, di sini kita memasukkan value andi....di posisi 1-5

plt.ylabel('ini y', size=20) #memberi label pada axis y
plt.yticks([1,2,3,4,5]) #mengatur value axis y

plt.title('ini judul', size=30) #memberi judul 
plt.show()
```

```python
# membuat lebih dari 1 plot dalam 1 canvas/grafik

plt.plot(x,y,'r--') #plot warna merah, bentuk garis2
plt.plot(y,x,'g*') #plot warna hijau, bentuk bintang2
plt.legend(['ini 1','ini 2']) #memberi legend
```

```python
plt.plot(x,y,'r')
plt.bar(x,y)
```

```python
# membuat subplot / 1 kanvas dipisah menjadi beberapa
plt.figure(figsize=(12,4))

plt.subplot(1,2,1) #baris, kolom, posisi # aku mau membuat subplot 1 baris 2 kolom, ini adalah subplot di posisi 1
plt.plot(x,y,'r--') #isi dari subplot 1

plt.subplot(1,2,2) # subplot di posisi 2
plt.plot(y,x,'g*') #isi dari subplot 2

```

```python
plt.figure(figsize=(12,10))

plt.subplot(221)
plt.plot(x,y,'r--')
plt.xlabel('ini x1')
plt.ylabel('ini y1')
plt.title('ini judul 1')

plt.subplot(222)
plt.plot(y,x,'g*')
plt.xlabel('ini x2')
plt.ylabel('ini y2')
plt.title('ini judul 2')

plt.subplot(223)
plt.plot(y,x,'b')
plt.xlabel('ini x3')
plt.ylabel('ini y3')
plt.title('ini judul 3')

plt.suptitle('ini judul utama', size=30)
# plt.tight_layout() #agar ukuran grafik pas dengan kanvas
plt.show()
```

## Artist Layer

```python
#buat kanvas
fig=plt.figure(figsize=(10,10))

# tambah axes
axes=fig.add_axes([0.1,0.1,0.5,0.5]) # membuat axes di posisi (kiri, bawah, lebar, tinggi (angka float 0-1))

#buat plot di axes
axes.plot(x,y,'b--')
axes.set_xlabel('ini sumbu x')
axes.set_ylabel('ini sumbu y')
axes.set_title('ini judul')
plt.show()
```

```python
#insert figure
# karena artist layer lebih flexibel
fig=plt.figure(facecolor='yellow') #face color: background color untuk 'pigura' gambar

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) # gambar2 yang lebih kecil di dalam gambar 1

axes1.plot(x,y)
axes1.set_xlabel('ini sumbu x1')
axes1.set_ylabel('ini sumbu y1')
axes1.set_title('ini judul 1')

axes2.plot(y,x,'r--')
axes2.set_xlabel('ini sumbu x2')
axes2.set_ylabel('ini sumbu y2')
axes2.set_title('ini judul 2')
plt.show()
```

```python
# subplotting di artist layer

fig,axes=plt.subplots(nrows=1, ncols=2, figsize=(8,4))

# axes# array berisi 2 data subplot matplotlib
axes[0].plot(y,x,'r--') #indexing untuk memilih posisi yang diinginkan
axes[1].plot(x,y)

for ax in axes:
    ax.set_xlabel('ini x')
    ax.set_ylabel('ini y')
    ax.set_title('ini judul')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2, figsize=(8,8))

# axes #array di dalam array
# axes[0] #array berisi subplot

axes[0][0].plot(x,y,'r--',label='ini 1')
axes[0][0].plot(y,x,'g*',label='ini 2')
axes[0][0].legend() #memberi legend, sesuai label yang diberikan di setiap plot
```
